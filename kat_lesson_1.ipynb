```python
import pandas as pd
```

```python
path = 'AppleStore.csv'
```

```python
apps = pd.read_csv(path)
```

```python
apps[0:5]
```

```python
#Create sub data set of 'Games' genre
game_apps = apps[[x for x in apps['prime_genre'] == 'Games']]
```

```python
len(game_apps)
```

```python
#Total number of game apps within each rating
game_apps.user_rating.value_counts().sort_values(ascending=False)
```

```python
#Find all unique values of cose
game_apps.price.unique()
```

```python
#show every game app that costs $3.99
cost_3_99 = game_apps[[x for x in game_apps['price'] == 3.99]]
cost_3_99
```
